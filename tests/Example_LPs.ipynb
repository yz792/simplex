```julia
# YIHAN: the commented out lines are the ones that I would like to run. 
# I'm runing with the others because they create smaller programs.
#xs = -6:0.01:6
xs = -6:0.5:6

#cs = 0.0:0.5:9
cs = 0.0:0.5:8

dist_cons = gauss_conv(xs, collect(cs), 1.0)
;
```

```julia
dist_cons
```

```julia
# YIHAN: I tried Double64 and BigFloat. 
# NT = Double64
NT = BigFloat
```

```julia
ONE = one(NT)
Z = zero(NT)
```

```julia
n = length(xs)
D1 = Variable(n)
T = Variable()
```

```julia
n
```

```julia
cons = D1 >= T
cons += sum(D1) == ONE
```

```julia
pairs = sym_pairs(xs)
for (i,j) in pairs
    cons += D1[i] == D1[j]
end
```

```julia
for k in dist_cons
    cons += k.rhs - dot(k.vec, D1) >= T
end
```

```julia
problem = maximize(T+2, cons, numeric_type = NT)
;
```

```julia
use default dual of primal constraint
```

```julia
solve!(problem, IterRef.Optimizer{NT}(verbose=true,mode=0))
```
